```python
import pandas as pd
import os

#  1. 추출할 컬럼 리스트 (ID 추가)
columns_to_extract = ['ID', '이용카드수_신용체크', '_1순위카드이용건수']

#  2. 파일 경로 리스트 생성 (201807 ~ 201812)
base_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/open/test/1.회원정보"
months = ['07', '08', '09', '10', '11', '12']
file_paths = [os.path.join(base_path, f"2018{month}_test_회원정보.parquet") for month in months]

#  3. 필요한 컬럼만 추출
df_list = []

for path in file_paths:
    df = pd.read_parquet(path)

    # 필요한 컬럼만 추출
    filtered_df = df[columns_to_extract].copy()

    # 기준년월 컬럼 추가
    기준년월 = os.path.basename(path).split('_')[0]
    filtered_df['기준년월'] = 기준년월

    df_list.append(filtered_df)

#  4. 병합
final_df = pd.concat(df_list, ignore_index=True)

#  5. 확인
display(final_df)
```

```python
import os

# 저장할 폴더 경로
save_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/Test회원정보_요약.csv"

# CSV 저장
final_df.to_csv(save_path, index=False, encoding='utf-8-sig')
```
