```python
import pandas as pd

# 1. 각 파일 경로
file_1_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/정보/Test/03.승인매출 정보1(test).csv"
file_2_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/정보/Test/3.승인매출2_test.csv"
file_3_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/정보/Test/청구잔액정보_병합(test).csv"
file_4_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/정보/Test/회원정보_요약(test).csv"

# 2. 파일 읽기
df1 = pd.read_csv(file_1_path)
df2 = pd.read_csv(file_2_path)
df3 = pd.read_csv(file_3_path)
df4 = pd.read_csv(file_4_path)

# 3. 병합 기준 키 확인
merge_keys = ['ID', '기준년월']

# 5. 병합 시도
merged1 = df1.merge(df2, on=merge_keys, how='inner')

# df3 병합
merged2 = merged1.merge(df3, on=merge_keys, how='inner')

# df4 병합
final_merged = merged2.merge(df4, on=merge_keys, how='inner')

# 결과 확인
print("최종 병합된 shape:", final_merged.shape)
print(final_merged.head())


# 6. 결과 확인
print("병합된 shape:", merged1.shape)
print(merged1.head())
```
