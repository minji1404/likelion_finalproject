```python
import pandas as pd
import glob
import os

# 🔹 1. 처리할 파일들 경로 (예: parquet 파일 2개)
file_paths = [
    "C:/Users/HR/Desktop/workspace/파이널프로젝트/open/train/7.마케팅정보/201807_train_마케팅정보.parquet",
    "C:/Users/HR/Desktop/workspace/파이널프로젝트/open/test/7.마케팅정보/201807_test_마케팅정보.parquet"
]

# 🔹 2. 병합용 리스트
df_list = []

for path in file_paths:
    df = pd.read_parquet(path)
    기준년월 = os.path.basename(path).split('_')[0]
    df['기준년월'] = 기준년월
    df_list.append(df)

# 🔹 3. 병합
merged_df = pd.concat(df_list, ignore_index=True)

# 🔹 4. 명세서 구성
desc_list = []

for col in merged_df.columns:
    if col == 'SEGMENT':
        continue  # ❗ 제외할 컬럼

    col_type = str(merged_df[col].dtype)
    nulls = merged_df[col].isnull().sum()
    nunique = merged_df[col].nunique()
    var_type = '연속형' if col_type in ['int64', 'float64'] else '범주형'

    # 이상치 계산 (연속형만)
    if var_type == '연속형':
        Q1 = merged_df[col].quantile(0.25)
        Q3 = merged_df[col].quantile(0.75)
        IQR = Q3 - Q1
        outliers = merged_df[
            (merged_df[col] < Q1 - 1.5 * IQR) | (merged_df[col] > Q3 + 1.5 * IQR)
        ].shape[0]
    else:
        outliers = 'N/A'

    # 예시값: 결측치 제외하고 고유값 상위 5개
    example_value = list(merged_df[col].dropna().unique()[:5])

    desc_list.append({
        '컬럼명': col,
        '데이터 타입': col_type,
        '결측치 수': nulls,
        '고유값 수': nunique,
        '이상치 수': outliers,
        '변수 유형': var_type,
        '예시값': example_value,
        '처리 계획': ''
    })

# 🔹 6. 명세서 저장
desc_df = pd.DataFrame(desc_list)
save_path = "C:/Users/HR/Desktop/workspace/파이널프로젝트/마케팅정보_명세서_201807_201807.xlsx"
desc_df.to_excel(save_path, index=False)

print(f"✅ 명세서 저장 완료!\n📄 위치: {save_path}")
```
