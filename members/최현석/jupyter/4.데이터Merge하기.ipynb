팀원 별로 담당했던 데이터를 하나로 합치기(Train)

```python
import pandas as pd
import glob
import os

#train파일 합치기

# 1) CSV 파일들이 들어 있는 디렉토리
train_dir = '/content/drive/MyDrive/멋사_파이널프로젝트/train'

# 2) 모든 CSV 경로 수집
csv_paths = glob.glob(os.path.join(train_dir, '*.csv'))

# 3) 순차적으로 병합
merged_df = None
for path in csv_paths:
    df = pd.read_csv(path)
    # 첫 파일이면 그대로 할당, 아니면 ID+기준년월 기준으로 병합
    if merged_df is None:
        merged_df = df
    else:
        merged_df = pd.merge(
            merged_df,
            df,
            on=['ID', '기준년월'],
            how='outer'    # 공통 key만 남기려면 'inner' 로 변경
        )

# 4) 결과 확인
print("최종 병합된 DataFrame 크기:", merged_df.shape)
display(merged_df.head())

# 5) 결과 저장 (원하는 경로로 변경)
out_path = os.path.join(train_dir, '/content/drive/MyDrive/파이널프로젝트 파일/merged_all_train.csv')
merged_df.to_csv(out_path, index=False, encoding='utf-8-sig')
print(f"✅ 모든 CSV 병합 완료: {out_path}")
```

팀원 별로 담당했던 데이터를 하나로 합치기(Test)

```python
# test파일 합치기

# 1) 테스트 CSV들이 들어 있는 디렉토리
test_dir = '/content/drive/MyDrive/멋사_파이널프로젝트/test'

# 2) 모든 CSV 경로 수집
csv_paths = glob.glob(os.path.join(test_dir, '*.csv'))
if not csv_paths:
    raise ValueError(f"No CSV files found in {test_dir}")

# 3) 순차적으로 병합 (ID, 기준년월 기준)
merged_test_df = None
for path in csv_paths:
    df = pd.read_csv(path)
    if merged_test_df is None:
        merged_test_df = df.copy()
    else:
        merged_test_df = pd.merge(
            merged_test_df,
            df,
            on=['ID', '기준년월'],
            how='outer'   # 공통 key만 남기려면 'inner'
        )

# 4) 결과 확인
print("최종 테스트 병합된 DataFrame 크기:", merged_test_df.shape)

# 5) 저장할 디렉토리 & 파일 경로
out_dir = '/content/drive/MyDrive/파이널프로젝트 파일'
os.makedirs(out_dir, exist_ok=True)
out_path = os.path.join(out_dir, 'merged_all_test.csv')

# 6) 병합된 결과 저장
merged_test_df.to_csv(out_path, index=False, encoding='utf-8-sig')
print(f"✅ 테스트 CSV 병합 완료: {out_path}")
```
