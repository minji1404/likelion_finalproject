```python
import pandas as pd

# 📁 기본 경로
base_path = r"C:\Users\user\Desktop\빅데이터 분석 부트캠프\파이널프로젝트\train"

# 📄 파일 불러오기
df1 = pd.read_csv(f"{base_path}\\03.승인매출 정보(train).csv")
df2 = pd.read_csv(f"{base_path}\\3.승인매출2_train.csv")
df3 = pd.read_csv(f"{base_path}\\청구잔액정보_병합(train).csv")
df4 = pd.read_csv(f"{base_path}\\Train회원정보2_요약.csv")

# 📌 기준년월 문자형(str)으로 통일
for df in [df1, df2, df3, df4]:
    if '기준년월' in df.columns:
        df['기준년월'] = df['기준년월'].astype(str)

# 🔗 병합 순서 (모두 ID + 기준년월 기준)
merged = pd.merge(df1, df2, on=['ID', '기준년월'], how='outer')
merged = pd.merge(merged, df3, on=['ID', '기준년월'], how='outer')
merged = pd.merge(merged, df4, on=['ID', '기준년월'], how='outer')

# ✅ 결과 확인
print("✅ 최종 병합 데이터 shape:", merged.shape)
print(merged.head())
```

```python
# 💾 CSV로 저장
merged.to_csv(f"{base_path}\\train.csv", index=False, encoding='utf-8-sig')
```
