```python
# 기본
import pandas as pd
import numpy as np
```

```python
# parquet 파일 데이터를 읽어온다.
train_df = pd.read_parquet('open/train/4.청구입금정보/201807_train_청구정보.parquet')
train_df
```

```python
# parquet 파일 데이터를 읽어온다.
test_df = pd.read_parquet('open/test/4.청구입금정보/201807_test_청구정보.parquet')
test_df
```

```python
# 두 데이터 프레임을 합친다.
all_df = pd.concat([train_df, test_df])
all_df.reset_index(inplace=True, drop=True)
all_df
```

```python
# 수치형/범주형 여부
for col in all_df.columns:
    dtype = all_df[col].dtype
    unique_vals = all_df[col].nunique()
    nulls = all_df[col].isnull().sum()
    
    print(f"\n📌 컬럼: {col}")
    print(f" - 데이터 타입: {dtype}")
    print(f" - 고유값 개수: {unique_vals}")
    print(f" - 결측치 수: {nulls}")
    
    if dtype == 'object' or unique_vals < 20:
        print(f" - 추정 유형: 범주형")
        print(f" - 예시값: {all_df[col].unique()[:5]}")
    elif pd.api.types.is_numeric_dtype(all_df[col]):
        print(f" - 추정 유형: 수치형")
        print(f" - 예시값: {all_df[col].dropna().unique()[:5]}")
```

```python
import pandas as pd

summary = []

for col in all_df.columns:
    series = all_df[col]
    dtype = str(series.dtype)
    unique_vals = series.nunique()
    nulls = series.isnull().sum()
    example_vals = series.dropna().unique()[:5]
    example_str = ', '.join(map(str, example_vals))

    # 범주형 여부
    is_categorical = (series.dtype == 'object') or (unique_vals < 20)

    # 수치형 여부 및 이산/연속 구분
    is_numeric = pd.api.types.is_numeric_dtype(series)
    is_discrete = (
        is_numeric and
        all(series.dropna().apply(lambda x: float(x).is_integer())) and
        unique_vals < 20
    )
    is_continuous = is_numeric and not is_discrete

    # 이상치 계산
    outlier_count = 0
    if is_numeric:
        q1 = series.quantile(0.25)
        q3 = series.quantile(0.75)
        iqr = q3 - q1
        lower = q1 - 1.5 * iqr
        upper = q3 + 1.5 * iqr
        outlier_count = series[(series < lower) | (series > upper)].count()

    summary.append({
        '컬럼명': col,
        '데이터 타입': dtype,
        '예시값': example_str,
        '범주형 (이산형)': '예' if is_categorical or is_discrete else '아니오',
        '수치형 (연속형)': '예' if is_continuous else '아니오',
        '결측치 수': nulls,
        '이상치 수': outlier_count
    })

summary_df = pd.DataFrame(summary)
summary_df.to_excel('데이터_명세서_04.청구입금정보_201807.xlsx', index=False)
```

```python
import pandas as pd

outlier_summary = []

for col in all_df.columns:
    if pd.api.types.is_numeric_dtype(all_df[col]):
        series = all_df[col].dropna()
        Q1 = series.quantile(0.25)
        Q3 = series.quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        outliers = series[(series < lower_bound) | (series > upper_bound)]

        outlier_summary.append({
            '컬럼명': col,
            '전체값 수': len(series),
            '이상치 수': len(outliers),
            '이상치 비율(%)': round(len(outliers) / len(series) * 100, 2),
            '최소값': series.min(),
            '최대값': series.max(),
            '이상치 기준 하한': round(lower_bound, 2),
            '이상치 기준 상한': round(upper_bound, 2)
        })

# 결과를 DataFrame으로 정리
outlier_df = pd.DataFrame(outlier_summary)
outlier_df
```
